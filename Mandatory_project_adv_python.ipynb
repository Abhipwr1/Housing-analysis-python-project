```python
def total_bedroom_size(total_bedroom):
  if total_bedroom<=10:
    return 'small'
  elif total_bedroom>=11 and total_bedroom<1000:
    return 'medium'
  else:
    return 'large'

df['total_bedroom_size']=df['total_bedrooms'].apply(total_bedroom_size)

df.head()
```

```python
from numpy.core.fromnumeric import sort
#6.median value of dataset
def median_value(data,column):
  if column in data:
    column_data=data[column]
    sort_data=sorted(column_data)

    n=len(sort_data)
    if n%2==0:
      med1=sort_data[n//2]
      med2=sort_data[n//2-1]
      median=(med1+med2)/2

    else:
      median=sort_data[n//2]

    return median
```

```python
med=median_value(df,'median_income')
print(med)
```
